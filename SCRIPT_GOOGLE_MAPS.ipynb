```python
import requests
import time

def obtener_detalles_lugar(place_id, api_key):
    url = "https://maps.googleapis.com/maps/api/place/details/json"
    params = {
        "place_id": place_id,
        "key": api_key,
        "fields": "name,rating,reviews,formatted_address"  # Asegúrate de incluir "reviews"
    }
    all_reviews = []
    next_page_token = None

    while True:
        if next_page_token:
            params['pagetoken'] = next_page_token
            time.sleep(2)  # Esperar un par de segundos antes de hacer la siguiente solicitud

        response = requests.get(url, params=params)
        if response.status_code == 200:
            data = response.json()
            if 'result' in data and 'reviews' in data['result']:
                all_reviews.extend(data['result']['reviews'])

            next_page_token = data.get('next_page_token')
            if not next_page_token:
                break
        else:
            print(f"Error al obtener datos: {response.status_code}")
            break

    return all_reviews

# Ejemplo de uso
place_id = "ChIJsX6q1bXIBZERqiIeSAseBhY"
api_key = GOOGLE_API_KEY

# Llamada a la función
reviews = obtener_detalles_lugar(place_id, api_key)

# Verifica si se obtuvieron las reseñas correctamente
if reviews:
    print(f"Total de reseñas obtenidas: {len(reviews)}")
    for review in reviews:
        print(review)
else:
    print("No se pudieron obtener las reseñas del lugar.")
```

```python
import requests

def obtener_detalles_lugar(place_id, api_key):
    url = "https://maps.googleapis.com/maps/api/place/details/json"
    params = {
        "place_id": place_id,
        "key": api_key,
        "fields": "name,rating,reviews,formatted_address"  # Asegúrate de incluir "reviews"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if 'result' in data:
            return data['result']
        else:
            return None
    else:
        print(f"Error al obtener datos para {place_id}: {response.status_code}")
        return None

def obtener_detalles_varios_lugares(place_ids, api_key):
    detalles_lugares = {}
    for place_id in place_ids:
        detalles = obtener_detalles_lugar(place_id, api_key)
        if detalles:
            detalles_lugares[place_id] = detalles
    return detalles_lugares

# Ejemplo de uso
place_ids = ["ChIJsX6q1bXIBZERqiIeSAseBhY", "ChIJBeFXk223BZERI4AN4cS6QBU", "ChIJL2B8xVPPBZERPBrSAsxGeqU"]  # Reemplaza con tus place_id --plaza de armas,morro solar,cerro san cristobal
api_key = GOOGLE_API_KEY

# Llamada a la función para varios lugares
detalles_lugares = obtener_detalles_varios_lugares(place_ids, api_key)

# Verifica si se obtuvieron los detalles correctamente
if detalles_lugares:
    for place_id, detalles in detalles_lugares.items():
        print(f"Detalles para {place_id}:")
        print(detalles)
else:
    print("No se pudieron obtener los detalles de los lugares.")
```

```python
import pandas as pd
import requests
from langdetect import detect

# Define the search parameters
location = '-12.045951295587582, -77.03056459170315'  # Coordinates for Lima, Perú
radius = 5000  # Search radius in meters
keyword = 'tourist attraction'  # Change depending on the category

# Define the base URL for the Places API
search_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
details_url = "https://maps.googleapis.com/maps/api/place/details/json"

params = {
    'location': location,
    'radius': radius,
    'keyword': keyword,
    'key': 'GOOGLE_API_KEY'  # Replace with your actual API key
}

# Function to get place details and reviews
def get_place_details(place_id):
    details_params = {
        'place_id': place_id,
        'fields': 'reviews',
        'key': 'GOOGLE_API_KEY'
    }
    details_response = requests.get(details_url, params=details_params)
    details_result = details_response.json().get('result', {})
    return details_result.get('reviews', [])

# Function to handle paginated results
def fetch_places(params):
    data = []
    reviews = []
    while True:
        response = requests.get(search_url, params=params)
        results = response.json().get('results', [])
        next_page_token = response.json().get('next_page_token', None)

        print(f"Number of results: {len(results)}")
        print(f"Next page token: {next_page_token}")

        for result in results:
            place = {
                'name': result.get('name'),
                'business_status': result.get('business_status'),
                'lat': result['geometry']['location']['lat'],
                'lng': result['geometry']['location']['lng'],
                'rating': result.get('rating'),
                'user_ratings_total': result.get('user_ratings_total'),
                'vicinity': result.get('vicinity'),
                'price_level': result.get('price_level'),
                'place_id': result.get('place_id'),
            }
            data.append(place)

            # Fetch place details to get reviews
            place_reviews = get_place_details(place['place_id'])
            for review in place_reviews:
                review_text = review.get('text', '')
                if review_text:
                    try:
                        language = detect(review_text)
                        print(f"Detected language: {language} for review: {review_text}")
                        if language == 'es':
                            reviews.append({
                                'place_id': place['place_id'],
                                'author_name': review.get('author_name'),
                                'rating': review.get('rating'),
                                'text': review_text,
                                'time': review.get('time')
                            })
                    except Exception as e:
                        print(f"Error detecting language for review: {review_text}, error: {e}")

        if next_page_token:
            params['pagetoken'] = next_page_token
            # Esperar unos segundos antes de realizar la siguiente solicitud debido a limitaciones de la API de Google Places
            import time
            time.sleep(2)
        else:
            break

    return data, reviews

# Fetch data
data, reviews = fetch_places(params)

# Create DataFrames
df_places = pd.DataFrame(data)
df_reviews = pd.DataFrame(reviews)

# Show the DataFrames
print(df_places)
print(df_reviews)
```
